```python
import pandas as pd

def clean_books(df):
    df = df.dropna(subset=["title", "description"])
    df["description"] = df["description"].str.lower()
    df["description"] = df["description"].str.replace(r"[^a-zA-Z ]", "", regex=True)
    return df
```
